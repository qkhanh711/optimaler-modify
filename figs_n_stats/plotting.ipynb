```python
import matplotlib.pyplot as plt
import json
import os
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
def revenue_viz(model_type):
    with open(cwd + f"/stats/{model_type}/official.json", "r") as f:
        train = json.load(f)
        profit = train["revenue"]
        plt.xlabel("Epochs")
        plt.ylabel("Revenue")
        plt.plot(profit, label="5x3")
        plt.legend()
        plt.show()
        plt.savefig(cwd + f"/figs/5x3_{model_type}_revenue.png")
```

```python
revenue_viz("RegretNet")
```

```python
def regret_viz(model_type):
    with open(cwd + f"/stats/{model_type}/official.json", "r") as f:
        train = json.load(f)
        regret = train["regret"]
        plt.xlabel("Epochs")
        plt.ylabel("Regret")
        plt.plot(regret, label="5x3")
        plt.legend()
        plt.show()
        plt.savefig(cwd + f"/figs/5x3_{model_type}_regret")
```

```python
regret_viz("RegretNet")
```

```python
revenue_viz("RegretFormer")
```

```python
regret_viz("RegretFormer")
```

```python
def spec_revenue_viz(model_type, num_agents, num_items):
    with open(cwd + f"/stats/{model_type}/specval_{num_agents}x{num_items}.json", "r") as f:
        train = json.load(f)
        profit = train["revenue"]
        plt.xlabel("Epochs")
        plt.ylabel("Revenue")
        plt.plot(profit, label=f"{num_agents}x{num_items}")
        plt.legend()
        plt.show()
        plt.savefig(cwd + f"/figs/{num_agents}x{num_items}_{model_type}_specval_revenue.png")
```

```python
def spec_regret_viz(model_type, num_agents, num_items):
    with open(cwd + f"/stats/{model_type}/specval_{num_agents}x{num_items}.json", "r") as f:
        train = json.load(f)
        regret = train["regret"]
        plt.xlabel("Epochs")
        plt.ylabel("Regret")
        plt.plot(regret, label=f"{num_agents}x{num_items}")
        plt.legend()
        plt.show()
        plt.savefig(cwd + f"/figs/{num_agents}x{num_items}_{model_type}_specval_regret")
```

```python
spec_revenue_viz("RegretFormer", 7, 3)
```

```python
spec_regret_viz("RegretFormer", 7, 3)
```

```python
spec_revenue_viz("RegretNet", 5, 3)
```

```python
spec_regret_viz("RegretNet", 5, 3)
```
