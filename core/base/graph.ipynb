```python
# Adjusting labels to avoid LaTeX parsing issues
plt.figure(figsize=(8, 6))

x0, t0 = 2, 1

x = numpy.linspace(-5, 5, 1000)
t = numpy.linspace(-5, 5, 1000)
t1 = x + t - 1
t2 = x + 2 * t - 4

# Characteristic lines
plt.plot(x, t1, "r", label=r"$\eta = t + x$")
plt.plot(x, t2, "b", label=r"$\xi = t - 2x$")

# Highlight the point (x0, t0)
plt.plot(x0, t0, "ko", label="Point (2, 1)")

# Fill domain of dependence and influence
plt.fill_between(
    x,
    t1,
    t2,
    where=(x >= x0 - 1) & (x <= x0),
    color="grey",
    alpha=0.3,
    label="Domain of Dependence",
)
plt.fill_between(
    x,
    t1,
    t2,
    where=(x >= x0) & (x <= x0 + 2),
    color="orange",
    alpha=0.3,
    label="Domain of Influence",
)

# Plot settings
plt.xlabel("x")
plt.ylabel("t")
plt.title("Domain of Dependence and Domain of Influence")
plt.xlim(-1, 5)
plt.ylim(-1, 4)
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
