```python
import json
import matplotlib.pyplot as plt


with open(path_trp3[0], "r") as f:
    results5x3_regretformer = json.load(f)['revenue']
with open(path_trp3[1], "r") as f:
    results5x3_regretnet = json.load(f)['revenue']

with open(path_trp4[0], "r") as f:
    results7x3_regretformer = json.load(f)['revenue']
with open(path_trp4[1], "r") as f:
    results7x3_regretnet = json.load(f)['revenue']

# Setting figure size and creating plot
figsize = (5, 4)
fig = plt.figure(figsize=figsize)
ax = fig.gca()

# Plotting the data
ax.plot(results5x3_regretformer, color="red", label="Attention-based ($trp^0= 10^{-3}$)")
ax.plot(results5x3_regretnet, color="green", label="FFN-based ($trp^0= 10^{-3}$)")
ax.plot(results7x3_regretformer, color="tomato", linestyle='--', label="Attention-based ($trp^0= 10^{-4}$)")
ax.plot(results7x3_regretnet, color="tab:green", linestyle='--', label="FFN-based ($trp^0= 10^{-3}$)")

# Adding labels and setting font size
ax.set_xlabel("No. of iterations", fontsize=15)
ax.set_ylabel("Revenue", fontsize=15)

# Setting ticks font size
ax.tick_params(axis='both', which='major', labelsize=18)

# Adding grid
ax.grid(True, linestyle='-', alpha=0.4)

# Adding legend
ax.legend(loc='lower right', fontsize=9.5, frameon=True, edgecolor='#333333')

# Saving and showing the figure
plt.savefig("trpRevenue.png", facecolor=fig.get_facecolor())
plt.show()
```

```python
num_iters = 300
results5x3_regretformer = [sum(results5x3_regretformer[i:i+num_iters])/num_iters for i in range(0, len(results5x3_regretformer), num_iters)]
results5x3_regretnet = [sum(results5x3_regretnet[i:i+num_iters])/num_iters for i in range(0, len(results5x3_regretnet), num_iters)]
results7x3_regretformer = [sum(results7x3_regretformer[i:i+num_iters])/num_iters for i in range(0, len(results7x3_regretformer), num_iters)]
results7x3_regretnet = [sum(results7x3_regretnet[i:i+num_iters])/num_iters for i in range(0, len(results7x3_regretnet), num_iters)]

print(f""" Revenue:
Attention-based (trp=10^3): {results5x3_regretformer[-1]}
FFN-based (trp=10^3): {results5x3_regretnet[-1]}
Attention-based (trp=10^4): {results7x3_regretformer[-1]}
FFN-based (trp=10^4): {results7x3_regretnet[-1]} 
""")
```
