```python
import matplotlib.pyplot as plt

# Data
false_rates = [0.4, 0.8, 1.2, 1.6]
percent_increase = [7.83031917608, 8.11949158631, 9.99421160922, 21.280985896]

# Create bar chart
plt.figure(figsize=(8,6))
bars = plt.bar(false_rates, percent_increase, color='green', width=0.15)

# Add line
plt.plot(false_rates, percent_increase, color='red', marker='o', linestyle='-', linewidth=2, markersize=5)

# Label the bars with percentages
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f"{int(yval)}%", ha='center', va='bottom', fontsize=15, color='black')

# Set the title and labels
# plt.title("Percent of increase of Revenue vs. False rate", fontsize=16)
plt.xlabel("False rate", fontsize=18)
plt.ylabel("Percent of increase of Payment", fontsize=18)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

# Adjust y-axis limit
plt.ylim(0, 35)

# Display grid
plt.grid(axis='y')

# Display the chart
plt.tight_layout()
plt.savefig("falsechart_with_line.pdf")
plt.show()
```
